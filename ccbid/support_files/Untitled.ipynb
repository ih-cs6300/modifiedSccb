```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA

train = pd.read_csv("~/Desktop/trainingD1.csv")
#test = pd.read_csv('~/Desktop/testingD1.csv')
answers = pd.read_csv('~/Desktop/species_idD1.csv')
```

```python
clf = RandomForestClassifier(max_depth=2, random_state=0)
```

```python
train
```

```python
pca = PCA(n_components=100, whiten=True)
pca.fit(train.iloc[:, 1:])
X = pca.transform(train.iloc[:, 1:])
```

```python
y = train.merge(answers, on='crown_id')['species_id']
y
```

```python
clf.fit(X, y)
```

```python
clf.score(X, y)
```
